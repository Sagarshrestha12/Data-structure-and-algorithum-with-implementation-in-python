```python
# Queue is similar as that of real life queue
# Queue is also the linear data structure in which operation are
# performed in the in FIFO order.
# In queue th insertion of the elements are done at the one end
# (rear) and deletion are performed at other end(front)
# Queue is same as real world queue.
# Operation on queue:
# Mainly the following four basic operations are performed on 
# queue: They are:
# Enqueue:Add an item to the queue.If the queue is full,it is 
# said to be Overflow condition.
# Dequeue:Removes the item from the queue.The items are poped in
# in the same order as they are pushed.If the stack is empty,
# then it is said to be an underflow condition
#
# Front: Get the front item of the queue
# Rear:Get the last item from the queue
# Queues maintain two data pointers,front and rear.
# for enqueue operation in stack:(add it to stack):
# rare =0, front=0; max size, x=[]
# void enqueue(sm): for enqueue
#     if(rare=maxsize):
#        print(queue is full)
#     else:
#       x.append(sm)
#       rare=rare++
# void dequeue():
#     if(rare==front):
#        print Queue is empty 
#     else:
#        y=x[rare-1]
#        
#       for i in range(0,rare-1)
#           x[i]=x[i+1]
#       #pointing rare to left by one decreasing 
#       rare -=1;
# void transverser():
#    for i in range(0,rare):
#        print(x[i])
# in queue two variable are used . one to abtain elts and 
#another to add the elts to queue
# First method to shift the elts towards left
class QueueExp(Exception):
    def __init__(self,x):
        print(x);
class queue:
    def __init__(self):
        self.a=[]
        self.max=5;
        self.rare,self.front=0,0;#initailly when the queue is empty both the front and rare point to same location
    #implementation of the enqueue i.e addition of the elts
    # to queue
    def ins(self,x): 
        if(self.rare==self.max):
            try:
                raise QueueExp("Queue is Full");
            except QueueExp:
                pass;
        else:
            self.a.append(x);
            self.rare += 1;
    def dequeue(self):
        if(self.rare==self.front):
            try:
                raise QueueExp("Queue is empty");
            except QueueExp:
                pass;
            #initially the
# front and rare variable point to 0 index
        else:
            y=self.a[0];# To get the elts from the bottom of
            # stack
            # now shifting elts form right to left one step
            for x in range(0,self.rare-1):
                self.a[x]=self.a[x+1]
            self.rare-=1;
            return y;
    def transverse(self):
        if(self.rare==self.front):
            print("Queue is empty so nothing to print");
        else:
            for x in range(0,self.rare):
                print(x);
x=queue();
for y in range(0,x.max):
    x.ins(y**2);
x.ins(34)
for y in range(x.max,0,-1):
    print(x.dequeue())
x.dequeue();
```

```python
class Computer:

    def __init__(self):
        self.__maxprice = 900

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

c = Computer()
c.sell()

# change the price
c.__maxprice = 1000
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```
