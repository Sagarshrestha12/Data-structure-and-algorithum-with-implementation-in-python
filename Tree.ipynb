```python
# Deletion operation in BST

# Algorihtum and its implemention in deletion operation of BST
def deletion(root,data):
    if root==None:
        return None
    elif(data>root.data):
        root.right=deletion(root.right,data)
    else:# if data== root.data then this block of code get 
        #executed
        if root.left==None:
            temp=root.right
            del root
            return temp
        elif root.right==None:
            temp=root.left
            del root
            return temp
        x=min(root.right)
        root.data=x
        root.right=deletion(root.right,x)
    return root;
```

```python
# Algorithum for finding the data in node
def search(node,data): #where data is to be found
        if node==None:
            return 0
        else:
            while(node.data!=data):
                if node.data<data:
                    node=node.right
                else:    
                    node=node.left
                if node==None:
                    return 0
                
                    
        return node;
```

```python
# Fucntion to check wheather a binary tree is Binary serch tree or not
min1=-783287
max2=4323423
# Making utility funtion
def isBst(node):
    return isBst2(node,min1,max2)

def isBst2(node,min1,max2): 
      
    # An empty tree is BST 
    if node is None: 
        return True
  
    # False if this node violates min/max constraint 
    if node.data < min1 or node.data > max2: 
        return False
  
    # Otherwise check the subtrees recursively 
    # tightening the min or max constraint 
    return (isBst2(node.left, min1, node.data -1) and
          isBst2(node.right, node.data+1, max2))   
```

```python
# Algorithum to find the height of the any tree
# Height of the empty tree is -1 ,we use this base condition
# in program 
# if tree is empty then its height is -1 
# and we divide huge problem into small problem using the 
# recursion since method of finding the height is same for
# small as well huge tree with large no. of node
# height of the tree is maximum number of the edges from the 
# root to leaf node
# we find the height of left subtree and right subtree and
# find which value is high and we add 1 to it and return as 
# the height of the node
def height(root):
    if root==None:
        return -1#As the height of the tree with no node is -1
    lsthgt=height(root.left)
    rsthgt=height(root.right)
    return 1 + (lsthgt if lsthgt>rsthgt else rsthgt)
# Notes:
# Height of tree always depends upon how we order instance 
# to design a tree
# Here the time complexity is O(n) where n is number of nodes
# As all nodes are visited at least once
```

```python
# Algorithum to find the inorder successor of the node
# in BST
def Successor(root,data):
    if root==None:
        return None;
    current=search(root,data);
    if (current==None):
        return None 
    elif(current.right!= None):
        return Tree().min(current.right)
    else:
        sucess=None
        ansestor=root
        while(ansestor!=current):
            if current.data<ansestor.data:
                sucess=ansestor
                ansestor=ansestor.left
            else:
                ansestor=ansestor.right
        return sucess
```
